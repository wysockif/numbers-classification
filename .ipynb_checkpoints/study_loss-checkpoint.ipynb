```python
mnist = tf.keras.datasets.mnist
(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train = tf.keras.utils.normalize(x_train, axis=1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
```

```python
def print_what_is_wrong(model):
    for i in range (0,100):
        pred = model.predict(x_test[i].reshape(1, 28, 28, 1))
        if pred.argmax() != y_test[i]:
            print(i, end=" ")
```

```python
print_what_is_wrong(model)
```

```python
def plot_loss_and_metric(val_loss, val_acc, loss, acc, metricname='accuracy'):
    plt.plot(val_acc)
    plt.title('model ' + metricname + ' zbioru testowego')
    plt.ylabel(metricname)
    plt.xlabel('epoch')
    plt.show()
    plt.plot(loss)
    plt.title('model loss zbioru uczacego')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.show()
```

```python
plot_loss_and_metric(val_loss, val_acc,loss,acc)
```
